```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
Radiologists_Assessment=label.fit_transform(data["Radiologists_Assessment"])
Is_Binary_Indicator_Of_Cancer_Diagnosis=label.fit_transform(data["Is_Binary_Indicator_Of_Cancer_Diagnosis"])
Comparison_Mammogram_From_Mammography=label.fit_transform(data["Comparison_Mammogram_From_Mammography"])
Patients_BI_RADS_Breast_Density=label.fit_transform(data["Patients_BI_RADS_Breast_Density"])
Family_History_Of_Breast_Cancer=label.fit_transform(data["Family_History_Of_Breast_Cancer"])
Current_Use_Of_Hormone_Therapy=label.fit_transform(data["Current_Use_Of_Hormone_Therapy"])
Binary_Indicator=label.fit_transform(data["Binary_Indicator"])
History_Of_Breast_Biopsy=label.fit_transform(data["History_Of_Breast_Biopsy"])
Is_Film_Or_Digital_Mammogram=label.fit_transform(data["Is_Film_Or_Digital_Mammogram"])
Cancer_Type=label.fit_transform(data["Cancer_Type"])
```

```python
data["Radiologists_Assessment"]=Radiologists_Assessment
data["Is_Binary_Indicator_Of_Cancer_Diagnosis"]=Is_Binary_Indicator_Of_Cancer_Diagnosis
data["Comparison_Mammogram_From_Mammography"]=Comparison_Mammogram_From_Mammography
data["Patients_BI_RADS_Breast_Density"]=Patients_BI_RADS_Breast_Density
data["Family_History_Of_Breast_Cancer"]=Family_History_Of_Breast_Cancer
data["Current_Use_Of_Hormone_Therapy"]=Current_Use_Of_Hormone_Therapy
data["Binary_Indicator"]=Binary_Indicator
data["History_Of_Breast_Biopsy"]=History_Of_Breast_Biopsy
data["Is_Film_Or_Digital_Mammogram"]=Is_Film_Or_Digital_Mammogram
data["Cancer_Type"]=Cancer_Type
```

```python
x=data.drop(["Radiologists_Assessment"],axis=1)
y=data["Radiologists_Assessment"]
```

```python
result=[]
models={'logistic':LogisticRegression(),'decision':DecisionTreeClassifier(),'random':RandomForestClassifier()}
for i in models:
    model=models[i]
    scores=cross_val_score(model,x,y,cv=5)
    a=(i,scores.mean())
    result.append(a)
```

```python
result
```

```python
models1={'gaussian':GaussianNB(),'bernoulli':BernoulliNB(),'multinomial':MultinomialNB()}
for i in models1:
    model=models1[i]
    scores=cross_val_score(model,x,y,cv=5)
    b=(i,scores.mean())
    result.append(b)
```
